```python
#Hierarchical Indexing means a DataFrame has more than one index level (multiple indexes for rows or columns).-
```

```python
import pandas as pd

data = {
    "Region": ["North", "North", "North", "South", "South", "South"],
    "Department": ["IT", "HR", "Sales", "IT", "HR", "Sales"],
    "Year": [2023, 2023, 2023, 2023, 2023, 2023],
    "Sales": [100000, 50000, 75000, 90000, 45000, 70000]
}

df = pd.DataFrame(data)

#df = df.set_index(["Region", "Department", "Year"])
print(df)
```

```python
#sets Region and Department as hierarchical (multi-level) row indexes
df.set_index(["Region", "Department"], inplace=True)
df
```

```python
#get north Region data with IT Department
df.loc[("North", "IT")]
```

```python
#Get North Region Data
df.loc["North"]
```

```python
tuples = [
    ("North", "IT"),
    ("North", "HR"),
    ("South", "IT"),
    ("South", "HR")
]
index = pd.MultiIndex.from_tuples(tuples, 
                                  names=["Region", "Department"])
df1= pd.DataFrame({"Sales": [100000, 50000, 90000, 45000]},index=index)
df1
```

```python
pop = [2020, 2021, 2022]
df2 = pd.DataFrame({'Year':pop})
df2
```
