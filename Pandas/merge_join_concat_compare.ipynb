```python
import pandas as pd
df = pd.DataFrame(
     {
         "foo": ["one", "one", "one", "two", "two", "two"],
         "bar": ["A", "B", "C", "A", "B", "C"],
         "baz": [1, 2, 3, 4, 5, 6],
         "zoo": ["x", "y", "z", "q", "w", "t"],
     }
 )
df1 = pd.DataFrame(
     {
         "foo": ["one1", "one1", "one1", "two1", "two1", "two1"],
         "bar": ["A1", "B1", "C1", "A1", "B1", "C1"],
         "baz": [1, 2, 3, 4, 5, 6],
         "zoo": ["x", "y", "z", "q", "w", "t"],
     }
 )
result = pd.concat([df1,df])
result
```

```python
#Join two data frames join="outerjoin"
pd.concat([df,df1],axis=1)
```

```python
#Inner join
result = pd.concat([df, df1], axis=1, join="inner")
result
```
