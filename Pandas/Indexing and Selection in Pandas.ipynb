```python
#Select Multiple Columns
df[["Name","Age"]]
```

```python
#Row Selection
df.iloc[0]
```

```python
#multiple ROw selection
df.iloc[0:2]
```

```python
#specific Column
df.loc[0:2, ["Name", "Salary"]]
```

```python
#specific Column
df.loc[:1, ["Name", "Salary"]]
```

```python
#Boolen indexing

df[df["Salary"]>55000]
```

```python
# 2. Salary with Department
df[(df["Salary"]>50000) & (df["Department"]=="HR")]
```

```python
#Query Method

df = pd.DataFrame({
    "Name": ["Amit", "Neha", "John", "Riya"],
    "Age": [25, 30, 28, 26],
    "Salary": [50000, 60000, 55000, 52000],
    "Department": ["IT", "HR", "IT", "Sales"]
})
```

```python
df.query("Department=='IT'")
```

```python
df.query("Salary > 50000 and Department=='IT'")
```

```python
#OR condition
df.query("Department == 'IT' or Department=='HR'")
```

```python
# Variable inside Query
value = 50000
df.query("Salary > @value")
```

```python
#Select Specific column after Query
value = 55000
df.query("Salary > @value")[["Name","Salary"]]
```
