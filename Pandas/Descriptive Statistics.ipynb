```python
import pandas as pd


data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [50000, 60000, 55000, 70000, 52000],
    'Experience': [2, 5, 3, 8, 4]
}
df = pd.DataFrame(data)
df
```

```python
df['Salary_Category'] = df['Salary'].apply(
    lambda x: 'High' if x > 60000 else 'Medium' if x >= 50000 else 'Low'
)
df['Exp_cat']=df['Experience'].apply(
    lambda x: 'Sr' if x >= 8 else 'Jr' if x >= 3 else 'Fresher'
)
df
```

```python
df['Promotion_Eligible'] = df.apply(
    lambda row: 'Yes' if (row['Salary'] > 50000) else 'No' , axis = 1
)
df
```

```python
## Conditional bonus calculation
df["Bonus"] = df.apply(
    lambda row: row["Salary"]*0.10 if(row["Department"]=='IT')
    else row["Salary"] * 0.08 if(row["Department"] == 'HR')
    else row["Salary"] * 0.12 ,
axis = 1
)
df
```
