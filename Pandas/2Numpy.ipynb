```python
import numpy as np
```

```python
#Normalize values between 0 and 1.
a = np.array([10, 20, 30, 40, 50])
normalized = (a - a.min()) / (a.max() - a.min())
print(normalized)
```

```python
#Find Missing Values
a = np.array([10, np.nan, 20, np.nan, 30])
np.isnan(a).sum()
```

```python
#replace missing value with mean
a[np.isnan(a)] = np.nanmean(a)
print(a)
```

```python
#Find top 3 highest values.
np.sort(a)[-3,]
```

```python
#Row-wise Mean in 2D Array
a = np.array([[10, 20], [30, 40]])

a.mean(axis=1)
```

```python
#column vise sum
a.sum(axis=0)
```

```python
#Get values greater than average.
a[a> a.mean()]
```

```python
#find index of max
np.argmax(a)
```

```python
#Standard Deviation Calculation
np.std(a)
```

```python
#Reshape Dataset for Model Input
np.reshape(-1,1)
```

```python
#Find Correlation-like Operation
a = np.array([10, 20, 30, 40])
b = np.array([15, 25, 35, 45])


np.corrcoef(a, b)
```
