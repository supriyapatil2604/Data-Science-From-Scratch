```python
import pandas as pd
import numpy as np
import statistics as stat

df = pd.read_csv("products-100000.csv")
print(df.describe())
```

```python
#Calculate with NumPy
#Range = Difference between maximum and minimum values.
#formula of data_range = max(data) - min(data)
data = df["Price"]
data_range = np.ptp(data)
print("Using Numpy",data_range)
```

```python
#Variance  = Avarage of Squared deviation from mean
#Using NumPy
variance = np.var(data)
print("Using Numpy",variance)

#Calculate with Statistics
variance_stat = stat.variance(data)
print("Using stat" , variance_stat)
```

```python
#Standard Deviation = Standard Deviation is a statistical measure that describes how much variation or dispersion there is in a set of data points.
std_dev= stat.stdev(data) 
print("Using stat" , std_dev)
std_dev_numpy=np.std(data)
print("Using NumPy", std_dev_numpy)
```

```python
# Interquartile Range
data_runs = {'runs': [10, 12, 13, 15, 16, 17, 20, 22, 90]}
data_runs_frame = pd.DataFrame(data_runs)

Q1 = data_runs_frame['runs'].quantile(0.25) #first quartile Formula: (n + 1) * 0.25
print("Q1 ",Q1)
Q3 = data_runs_frame['runs'].quantile(0.75) #second quartile
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

df_cleaned = data_runs_frame[(data_runs_frame['runs'] >= lower_bound) & (data_runs_frame['runs'] <= upper_bound)]
print("\n Cleaned data (After calculating of Outliers ):")
print(df_cleaned)
```

```python
# 1. Sample Data
data_runs = [10, 12, 13, 15, 16, 17, 20, 22, 90]

# 2. Sort the data (Crucial for Quartiles)
sorted_runs = sorted(data_runs)
n = len(sorted_runs)
# 3. Calculate the Position (Formula: (n + 1) * 0.25)
# Note: In Python, indexing starts at 0, so we adjust accordingly
pos = (n + 1) * 0.25
print("pos ",pos)

# 4. Extract Integer and Fractional Parts
# Example: If pos is 2.5, index is 2, fraction is 0.5
index = int(pos)
fraction = pos - index
print("index ", index)
print("fraction ", fraction)

# 5. Logic to find the value (Interpolation)
if fraction == 0:
    # If it's a whole number, take the value at that position
    q1 = sorted_runs[index - 1]
    print("q1 ",q1)

else:
    # Interpolate between the two nearest numbers
    # value1 + fraction * (value2 - value1)
    val1 = sorted_runs[index - 1]
    print("val1 ",val1)
    val2 = sorted_runs[index]
    print("val2 ",val2)

    q1 = val1 + fraction * (val2 - val1)
    print("q1 ",q1)

print(f"Sorted Data: {sorted_runs}")
print(f"Q1 Position: {pos}")
print(f"**Manual Q1 Result: {q1}**")
```
